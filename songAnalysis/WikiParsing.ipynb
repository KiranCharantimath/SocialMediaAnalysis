```python
import bs4 
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as Soup 
my_url="https://en.wikipedia.org/wiki/Sony_Music_India"
uClient=uReq(my_url)
page_html=uClient.read()
uClient.close()
page_soup=Soup(page_html,"html.parser")
```

```python
page_soup
```

```python
headline = page_soup.find_all(class_="mw-headline")
```

```python
films=[]
for head in headline:
    print(head["id"])
    films.append(head["id"])
films
```

```python
tables = page_soup.findAll("table",{"class":"multicol"})
```

```python
files=[]
for i in range(0,len(tables)):
    files.append(open(films[i]+".txt",'w'))

count=2
for i in range(0,len(tables)):
    count=2
    flag=0
    for j in tables[i]:
        for x in j.dl:
            if(count%2==0):
                print(j.getText())
                files[i].write(j.getText())
            count+=1
            break
    files[i].close()
    print("-------------------")
files[i-1].close()
            
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
 
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
 
    return False
sony_music_Tamil={}
sony_music_Tamil["year_of_release"]=[]
sony_music_Tamil["songs"]=[]
with open("Tamil_Films.txt") as f:
    for line in f:
        if line!="\n" and is_number(line) and int(line)!=3:
            print(line)
            sony_music_Tamil["year_of_release"].append(int(line.strip()))

f.close()
i=-1
with open("Tamil_Films.txt") as f:
    for line in f:
        if line!="\n":
            if is_number(line) and int(line)!=3:
                sony_music_Tamil["songs"].append([])
                i+=1
                continue
            sony_music_Tamil["songs"][i].append(line.strip())
            
f.close()
            
            
print(sony_music_Tamil)
```

```python
sony_music_Hindi={}
sony_music_Hindi["year_of_release"]=[]
sony_music_Hindi["songs"]=[]
with open("Hindi_Films.txt") as f:
    for line in f:
        if line!="\n" and is_number(line):
            print(line)
            sony_music_Hindi["year_of_release"].append(int(line.strip()))

f.close()
i=-1
with open("Hindi_Films.txt") as f:
    for line in f:
        if line!="\n":
            if is_number(line):
                sony_music_Hindi["songs"].append([])
                i+=1
                continue
            sony_music_Hindi["songs"][i].append(line.strip())
            
f.close()
            
            
print(sony_music_Hindi)
```
